# Pricing and Margin Analysis

This notebook evaluates pricing strategies and profit margins to optimize revenue and profitability.

Key goals:
- Assess pricing effectiveness
- Analyze margin distribution across categories
- Understand relationship between price, sales, and profit
- Identify opportunities for pricing optimization

Pricing analysis evaluates how product pricing impacts:

- Sales volume
- Revenue generation
- Profitability



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure datetime if needed
# orders['order_date'] = pd.to_datetime(orders['order_date'])

# Revenue (if not already present)
orders['revenue'] = orders['price'] * orders['quantity']

# Margin
orders['margin'] = orders['revenue'] - orders['cost']

# Margin %
orders['margin_pct'] = orders['margin'] / orders['revenue']
```

## Key Pricing Metrics

- Average Selling Price (ASP)
- Revenue
- Margin
- Margin %
- Discount Impact

```python
pricing_metrics = orders.agg({
    'price': 'mean',
    'revenue': 'sum',
    'margin': 'sum',
    'margin_pct': 'mean'
})

pricing_metrics
```

## Category-Level Analysis

We analyze how pricing and margins differ across product categories.

```python
category_analysis = orders.groupby('category').agg({
    'revenue': 'sum',
    'margin': 'sum',
    'margin_pct': 'mean',
    'price': 'mean'
}).sort_values(by='revenue', ascending=False)

category_analysis
```

## Margin Distribution

This shows how profitability varies across transactions.

```python
plt.figure(figsize=(8,5))
sns.histplot(orders['margin_pct'], bins=30)
plt.title("Margin % Distribution")
plt.show()
```

## Price vs Sales

Analyzing relationship between price and quantity sold.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=orders, x='price', y='quantity')
plt.title("Price vs Quantity Sold")
plt.show()
```

## Price vs Margin

Understanding how pricing impacts profitability.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=orders, x='price', y='margin_pct')
plt.title("Price vs Margin %")
plt.show()
```

## Discount Impact Analysis

Understanding how discounts affect revenue and margin.

```python
discount_analysis = orders.groupby('discount').agg({
    'revenue': 'mean',
    'margin_pct': 'mean'
})

discount_analysis
```

## Key Insights

- Higher prices do not always reduce sales (identify pattern)
- Some categories generate high revenue but low margin
- Discounts increase sales but reduce profitability
- Margin distribution shows variability across products

## Business Interpretation

- Pricing strategy may be volume-driven instead of profit-driven
- Some products may be underpriced
- Discount-heavy strategy may be eroding margins

## Recommendations

- Optimize pricing for low-margin high-volume products
- Reduce unnecessary discounts
- Focus on high-margin categories
- Implement dynamic pricing strategies
- Bundle products to increase average order value
